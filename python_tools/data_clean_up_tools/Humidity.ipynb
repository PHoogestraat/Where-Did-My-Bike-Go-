Humidity Clean up

Converts date into year month week day hour

```python
# Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn import tree
import os
```

```python
path = "data/Weather/"
file = "humidity.csv"
path_file = path + file
df = pd.read_csv(path_file)
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df['Date(YMD)'] = pd.to_datetime(df['datetime'])
```

```python
df['DATE_year'] = df['Date(YMD)'].dt.year
df['DATE_month'] = df['Date(YMD)'].dt.month
df['DATE_day'] = df['Date(YMD)'].dt.day
df['Hour'] = df['Date(YMD)'].dt.hour
df['Minute'] = df['Date(YMD)'].dt.minute

df
```

```python
presure_indy_simple_df = df.drop(columns=["Date(YMD)","datetime", "Minute"])
presure_indy_simple_df
```

```python
presure_indy_simple_df.rename(columns = {"Indianapolis": "Humidity%"}, inplace=True)
presure_indy_simple_df
```

```python
#      save to csv for other models
file = "IndyHumiditySpeed_2017_2012.csv"
path_file = path + file
presure_indy_simple_df.to_csv(path_file)
```
